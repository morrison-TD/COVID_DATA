```
frames=[df1,df3,df3]
merged_df=pd.concat(frames)
merged_df.info()
```

```
#dropped columns not needed
dropped_df = merged_df.drop(columns=['cover','Unnamed: 0','Unnamed: 9','added','rating.1'])
dropped_df.head()
```

```
#dropped 'author', convert to str
#dropped'avg rating' convert to float
#dropped 'read date' 
# dropped 'title' convert to str
dropped_df.author = dropped_df.author.str.replace("author","", regex=True).astype(str)
dropped_df.rating = dropped_df.rating.str.replace("avg rating","").astype(float)
dropped_df.read = dropped_df.read.str.replace("date read","", regex=True)
dropped_df.title = dropped_df.title.str.replace("title","", regex=True).astype(str)
dropped_df
#dropped_df.head()
```

```
dropped_df.read = dropped_df.read.str.replace("not set","", regex=True)
dropped_df
#dropped_df.head()
```

```
dropped_df['read'] =  pd.to_datetime(dropped_df['read'])
                                    #  , format='%d%b%Y:%H:%M:%S.%f')
dropped_df
```
